```python
import pandas as pd
```

```python
# Reload the Arizona data using the correct header row (index 4)
az_radiology_df = pd.read_csv("export_az_radiology.csv", skiprows=5)




# Drop rows without a CPT code
az_radiology_df = az_radiology_df[az_radiology_df['Code'].notna()]



# Rename and normalize columns
az_radiology_df = az_radiology_df.rename(columns={
    'Code': 'proc_cd',
    'FY24 FAC RBRVS Rate': 'rate'
})

# Add required fields
az_radiology_df['modifier'] = None
az_radiology_df['description'] = None
az_radiology_df['rate_unit'] = 1
az_radiology_df['is_by_report'] = False
az_radiology_df['region_type'] = 'state'
az_radiology_df['region_value'] = 'AZ'

# Select final columns
final_columns = [
    'proc_cd', 'modifier', 'description',
    'rate', 'rate_unit', 'is_by_report',
    'region_type', 'region_value'
]

az_radiology_final = az_radiology_df[final_columns]



# Split the proc_cd by space into base code and modifier
az_radiology_final[['proc_cd', 'modifier']] = az_radiology_final['proc_cd'].str.split(' ', n=1, expand=True)

# Treat "00" as no modifier
az_radiology_final['modifier'] = az_radiology_final['modifier'].replace('00', None)

# Reorder columns (in case they shifted)
final_columns = [
    'proc_cd', 'modifier', 'description',
    'rate', 'rate_unit', 'is_by_report',
    'region_type', 'region_value'
]

az_radiology_final = az_radiology_final[final_columns]

# Export cleaned version
az_output_cleaned_path = "db_import_arizona.csv"
az_radiology_final.to_csv(az_output_cleaned_path, index=False)

display(az_radiology_final.head())
```
